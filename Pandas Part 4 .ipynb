```python
import pandas as pd
import matplotlib.pyplot as plt
world = pd.read_csv('2018-2019 Happiness.csv') #Importing world hapiness dataset
airbnd = pd.read_csv('airbnb.csv')
salaries = pd.read_csv('salaries.csv')
websales= pd.read_csv('web_sales.csv')
eth = pd.read_csv('Ethereum Historical Data.csv', index_col= 0)
```

```python
eth.info()
```

```python
eth['Change %'] = eth['Change %'].replace('%','', regex = True)
```

```python
eth.info()
```

```python
websales['product_subcategory'] = websales['product_subcategory'].replace('caps','Hats')
```

```python
websales['SalesTerritoryRegion'] = websales['SalesTerritoryRegion'].replace(['Northwest', 'Southwest'], ['West USA', 'West USA'])
```

```python
websales['SalesTerritoryRegion'].value_counts()
```

```python
websales[' OrderDateKey '] = pd.to_datetime(websales['OrderDateKey'], format = '%Y%m%d')
```

```python
websales.groupby('CalendarYear')['SalesAmount'].max()
```

```python
salaries[salaries['job_title'] == 'Data Scientist']['salary_in_usd'].mean()
```

```python
salaries.groupby('job_title')['salary_in_usd'].max().sort_values(ascending=False)
```

```python
salaries.groupby(['job_title','company_location', 'experience_level'])['salary_in_usd'].mean().sort_values(ascending=False)
```

```python
salaries.groupby('job_title')['salary_in_usd'].agg(['mean', 'min', 'max']).sort_values(by = 'mean', ascending = False).head(25)
```

```python
salaries.groupby('job_title').agg({'salary_in_usd':['mean', 'min', 'max'],'remote_ratio':['mean']})
```

```python
salaries.groupby('job_title').agg(avg_salary = ('salary_in_usd','mean'),
                                  min_salary = ('salary_in_usd', 'min'),
                                  max_salary = ('salary_in_usd', 'max'))
```

```python
work_year_avg_salary_with_experience = salaries.groupby(['work_year', 'experience_level'])['salary_in_usd'].mean().unstack()
```

```python
work_year_avg_salary_with_experience.stack()
```

```python
work_year_avg_salary_with_experience.plot()
```

```python
work_year_avg_salary_with_experience.plot(kind = 'bar' , figsize = (10,6))
```
